# Covariance

## Definition
Covariance measures how two variables change together.
- If both increase together → positive covariance
- If one increases while the other decreases → negative covariance
- If there is no consistent pattern → covariance near zero

Covariance is not scaled and depends on the units of the variables.

# Correlation

## Definition
Correlation measures the strength and direction of a linear relationship between two variables.  
It is a scaled version of covariance and always lies between **−1 and 1**.

- +1 → perfect positive linear relationship  
- -1 → perfect negative linear relationship  
- 0 → no linear relationship  

Correlation is unit-free and easier to interpret than covariance.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

x = np.array([10,20,30,40,50])
y = np.array([23,34,45,67,89])

x_bar = np.mean(x)
y_bar = np.mean(y)

covariance = np.sum((x - x_bar) * (y - y_bar)) / (len(x) - 1)
print("Covariance:", covariance)

corr_matrix = np.corrcoef(x, y)
print("Pearson correlation matrix:\n", corr_matrix)

sns.heatmap(corr_matrix, annot=True, vmin=-1, vmax=1, center=0, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
