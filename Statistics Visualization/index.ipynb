## Descriptive Statistics

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate sample data
data = np.random.normal(loc=50, scale=10, size=200)

# Calculate statistics
mean = np.mean(data)
median = np.median(data)
std_dev = np.std(data)

print(f"Mean: {mean:.2f}, Median: {median:.2f}, Std Dev: {std_dev:.2f}")

# Visualize
sns.histplot(data, kde=True, color='skyblue')
plt.axvline(mean, color='red', linestyle='--', label='Mean')
plt.axvline(median, color='green', linestyle='--', label='Median')
plt.legend()
plt.title("Data Distribution with Mean and Median")
plt.show()
```

### Probability Distribution

```python
from scipy.stats import uniform, binom, norm

x = np.linspace(-5, 5, 100)

# Normal Distribution
plt.plot(x, norm.pdf(x, 0, 1), label='Normal(0,1)')
# Uniform Distribution
plt.plot(x, uniform.pdf(x, -2, 4), label='Uniform(-2,2)')
plt.legend()
plt.title("Probability Density Functions")
plt.show()

# Binomial Distribution
n, p = 10, 0.5
k = np.arange(0, n+1)
plt.bar(k, binom.pmf(k, n, p))
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.show()
```

### Central Limit Theorem

```python
samples = [np.mean(np.random.uniform(0, 1, 100)) for _ in range(1000)]

sns.histplot(samples, kde=True, color='orange')
plt.title("Central Limit Theorem: Sampling Distribution of the Mean")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```

### Correlation and Covariance

```python
# Generate correlated data
x = np.random.randn(100)
y = 2*x + np.random.randn(100)*0.5

sns.scatterplot(x=x, y=y)
plt.title("Positive Correlation Example")
plt.show()

corr = np.corrcoef(x, y)[0, 1]
print(f"Correlation coefficient: {corr:.2f}")
```
