### Assignment 15: Generator Pipeline

Write three generator functions: `integers` that yields integers from 1 to 10, `doubles` that yields each integer doubled, and `negatives` that yields the negative of each doubled value. Chain these generators to create a pipeline that produces the negative doubled values of integers from 1 to 10.

```python
class Countdown:
    def __init__(self, num):
        self.num = num

    def __iter__(self):
        return self

    def __next__(self):
        if self.num < 0:
            raise StopIteration
        current_num = self.num
        self.num -= 1
        return current_num

obj = Countdown(10)
for num in obj:
    print(num)
```

```python
## Ass2

class MyRange:
    def __init__(self, start, stop=None, step=1):
        if stop is None:
            start, stop = 0, start
        self.start = start
        self.stop = stop
        self.step = step
        self.current = start

    def __iter__(self):
        return self

    def __next__(self):
        if (self.step > 0 and self.current >= self.stop) or (self.step < 0 and self.current <= self.stop):
            raise StopIteration
        value = self.current
        self.current += self.step
        return value

# Test the class
for num in MyRange(5):
    print(num)

print("Range with start and stop:")
for num in MyRange(1, 5):
    print(num)

print("Range with start, stop, and step:")
for num in MyRange(1, 10, 2):
    print(num)

print("Range with negative step:")
for num in MyRange(10, 1, -2):
    print(num)
```

```python
## Ass3

def fibonacci():
    """Generator function for Fibonacci sequence"""
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b

# Test the generator
fib_gen = fibonacci()
for _ in range(10):
    print(next(fib_gen))
```
