### Reading HTML

```python
url="https://www.fdic.gov/bank-failures/failed-bank-list"
df = pd.read_html(url)
```

```python
df[0]
```

### If I want a particular table, Then I can use the `match` keyword in `read_html`   

```python
## Example 2

url2 = "https://en.wikipedia.org/wiki/Mobile_country_code"
df_mobile_code = pd.read_html(url2, match="Country", header=0)
df_mobile_code[0]
```
