### Data Aggregating and Grouping

```python
## Grouping based on country and AR Score

grouped_AR_Score = df.groupby('Country/Territory')['AR SCORE'].mean().sort_values()
print(grouped_AR_Score) 
```

```python
grouped_sum = df.groupby(['Country/Territory', 'Focus'])['AR SCORE'].sum()
print(grouped_sum)
```

### Aggregate multiple functions

```python
grouped_agg = df.groupby('Region')['AR SCORE'].agg(['mean', 'sum', 'count'])
print(grouped_agg)
```

### Merging and joining dataframes

```python
df1 = pd.DataFrame({'Key': ['A', 'B', 'C'], 'Value1': [1,2,3]})
df2 = pd.DataFrame({'Key': ['A', 'B', 'D'], 'Value1': [4,5,6]})
```

```python
df1
```

```python
df2
```

```python
## Merge dataframe on the 'Key' column

pd.merge(df1, df2, on="Key", how="inner")
```

```python
pd.merge(df1, df2, on="Key", how="left")
```

```python
pd.merge(df1, df2, on="Key", how="right")
```

```python
pd.merge(df1, df2, on="Key", how="outer")
```
