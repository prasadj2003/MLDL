### Seaborn is a Python data visualization library built on top of Matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## There are many datasets available inside seaborn
tips = sns.load_dataset('tips')
tips
```

```python
## Scatter plot

sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title("Total bll vs Tip")
plt.show()
```

```python
## Line plot

sns.lineplot(x='size', y='total_bill', data=tips)
plt.title("size vs total bill")
plt.show()
```

```python
### Categorical plots

# Bar plot

sns.barplot(x='day', y='total_bill', data=tips, color='orange')
plt.title("day wise total bill")
plt.show()
```

```python
## Box plot 

sns.boxplot(x='day', y='total_bill', data=tips, color='teal')
plt.title("day wise total bill box plot")
plt.show()

## The bubbles in below figures are outliers
```

```python
## To see the distribution of data We See the VIOLIN PLOT

sns.violinplot(x='day', y='total_bill', data=tips, color='pink')
plt.title("violin plot of day wise bill")
plt.show()
```

```python
## Histograms

sns.histplot(tips['total_bill'], bins=10, kde=True)
plt.title("histogram of total bill")
plt.show()

## If kde=False then we will not get the curve
```

```python
## KDE plot

sns.kdeplot(tips['total_bill'], fill=True)
```

```python
## Pair plot -> Used to view relationship between different tables

sns.pairplot(tips)
```

```python
## Heatmap for correlation

corr = tips[['total_bill', 'tip', 'size'] ].corr()
corr
```

```python
sns.heatmap(corr, annot=True, cmap='coolwarm')
```

```python
import pandas as pd
```

```python
sales_data = pd.read_excel('Adidas US Sales Datasets.xlsx')
sales_data.head()
```

```python
## claning the data

sales_data.drop(columns=['Unnamed: 0'], axis=1, inplace=True)
sales_data.drop([0,1,2], axis=0, inplace=True)

sales_data.head()
```

```python
## total sales by prodct

# converting total sales to float
sales_data['Unnamed: 10'] = pd.to_numeric(sales_data['Unnamed: 10'], errors='coerce')

sns.lineplot(x='Unnamed: 7', y='Unnamed: 10', data=sales_data, estimator=sum)
plt.tick_params(axis='x', rotation=90)
plt.xlabel("Product")
plt.ylabel("Sales")
plt.title("total sales by product")
plt.show()
```
