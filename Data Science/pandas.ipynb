Pandas is a powerful Python library for data manipulation and analysis. Here are some basic concepts:
Key Features:
- DataFrames: Two-dimensional labeled data structures with columns of potentially different types.
- Series: One-dimensional labeled array of values that can be used to store and manipulate data.

```python
import pandas as pd
```

```python
data = [19,24,32,4,5]
series = pd.Series(data)
print("series: \n", series)
```

```python
## Series from dictionary

data = {'s': 1, 'd':3, 't':9}
series_dict = pd.Series(data)
print(series_dict)
```

```python
data=[10,20,30]
index=['a', 'b', 'c']
pd.Series(data, index=index)
```

### Dataframes

```python
## create dataframe from dictionary of list

data = {
    'Name': ['Prasad', 'Rajan', 'Om'],
    'Age': [22,15,42],
    'City': ['Pune', 'NYC', 'Wai']
}

df = pd.DataFrame(data)
print(df)
print(type(df))
```

```python
## Df from list of dictionaries

data=[
    {'Name': 'Prasad', 'Age': 22, 'City': 'NYC'},
    {'Name': 'Rajan', 'Age': 22, 'City': 'Nebraska'},
    {'Name': 'Prasad', 'Age': 22, 'City': 'NYC'},
]

df = pd.DataFrame(data)
print(df)
print(type(df))
```
