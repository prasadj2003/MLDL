### Multiple plots

```python
## sample data

x = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [1,2,3,4,5]

plt.figure(figsize=(9,5)) # this is the parent figure, 9,5 w.r.t. pixels

plt.subplot(1,2,1) ## 1 rows and 2 columns AND third parameter is where I will plot the figure, in which quadrant?
plt.plot(x,y1, color='green')
plt.title("plot 1")

plt.subplot(1,2,2) ## 1 rows and 2 columns AND third parameter is where I will plot the figure, in which quadrant?
plt.plot(y1,x, color='red')
plt.title("plot 2")

plt.subplot(2,2,3) 
plt.plot(x,y2, color='blue')
plt.title("plot 3")

plt.subplot(2,2,4) 
plt.plot(y2,x, color='black')
plt.title("plot 4")
```

### Adidas US sales data visualization

```python
import pandas as pd
```

```python
sales_data = pd.read_excel("Adidas US Sales Datasets.xlsx")
sales_data.head(5)
```

```python
sales_data.columns
```

```python
## I have to drop rows 0-2 and column 0

## dropping columns
sales_data.drop(columns=['Unnamed: 0'], axis=1, inplace=True)

## dropping rows
sales_data.drop([0,1,2], axis=0, inplace=True)

sales_data.head(5)
```

```python
sales_data.dtypes
```

```python
## all are string or mix datatypes. We will use pd.to_numeric to convert them

sales_data['Unnamed: 10'] = pd.to_numeric(sales_data['Unnamed: 10'], errors='coerce')
sales_data['Unnamed: 11'] = pd.to_numeric(sales_data['Unnamed: 11'], errors='coerce')
sales_data.dtypes
```

```python
## Checking if Unnames: 10 and Unnamed: 11 have Nan values. If yes, replace them with mean values

print("null values in Unnamed: 10: ", sales_data['Unnamed: 10'].isnull().sum())
print("null values in Unnamed: 11: ", sales_data['Unnamed: 11'].isnull().sum())

sales_data['Unnamed: 10'] = sales_data['Unnamed: 10'].fillna(sales_data['Unnamed: 10'].mean())
sales_data['Unnamed: 11'] = sales_data['Unnamed: 11'].fillna(sales_data['Unnamed: 11'].mean())

print("null values in Unnamed: 10: ", sales_data['Unnamed: 10'].isnull().sum())
print("null values in Unnamed: 11: ", sales_data['Unnamed: 11'].isnull().sum())
```

```python
## total sales vs operating profit

plt.figure(figsize=(9,5))

plt.subplot(1,2,1)
plt.title("total sales vs operating profit")
plt.plot(sales_data['Unnamed: 10'], sales_data['Unnamed: 11'])
plt.xlabel("total sales")
plt.ylabel("Operating Profit")

plt.subplot(1,2,2)
plt.title("total sales vs operating profit")
plt.scatter(sales_data['Unnamed: 10'], sales_data['Unnamed: 11'], marker='.', color='green')
plt.xlabel("total sales")
plt.ylabel("Operating Profit")

```

```python
## Total sales by product

sales_data.groupby('Unnamed: 7')['Unnamed: 10'].sum().plot(kind='bar')
plt.title('Total Sales by Product')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.show()
```

```python
## Region wise total sale

Region_wise_sale = sales_data.groupby('Unnamed: 4')['Unnamed: 10'].sum()
Region_wise_sale
```

```python
Region_wise_sale.plot(kind='bar', color='Teal', xlabel='Region', ylabel='Total Sales')
```
